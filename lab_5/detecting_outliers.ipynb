```python
import numpy as np

import matplotlib.pyplot as plt

data = np.array([1383.464, 1384.9871, 1386.6281, 1388.142, 1389.766, \
                 1391.188, 1392.796, 1394.144, 1395.4139, 1396.488, \
                 1398.209, 1500.01106419, 1230.79299225, 1549.47510581, \
                 1399.6281, 1238.03303254, 1400.31845915, 1249.40048243, \
                 1400.764, 1363.80564471, 1356.90945608, 1287.4883591, \
                 1402.229, 1403.588, 1405.167, 1406.427])

def mad(data):
    median_data = np.median(data)
    d_data = [abs(x - median_data) for x in data]

    Tm_data = np.median(d_data)
    tolerance = Tm_data * 3
    return [x > tolerance for x in d_data]

x, y = zip(*enumerate(data))

plt.scatter(x, y, c=mad(data))
```
